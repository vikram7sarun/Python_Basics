```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = [2,6,4,5,2,8,3]
y = [21,45,36,73,91,33,27]
plt.scatter(x,y,color='green',marker='+')
plt.plot(x,y)
plt.xlabel("length")
plt.ylabel("height")
plt.title("Plot Chart")
plt.plot(x, y, label='linear')
plt.legend()
plt.set_alpha(x,y)
```
